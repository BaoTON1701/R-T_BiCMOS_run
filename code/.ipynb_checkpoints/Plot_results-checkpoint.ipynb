## Input noise plot 

```python

col = ['f', 'LNA1', 'LNA2']
input_noise.columns = col

plt.figure(figsize=(10, 6), layout="constrained")
plt.loglog(input_noise['f'], input_noise['LNA1'], label='LNA1', color=colors[1])
plt.loglog(input_noise['f'], input_noise['LNA2'], label='LNA2', color=colors[2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Input Referred Noise (nV/√Hz)')
plt.title('Input Referred Noise vs Frequency')
plt.grid(True, which='both', ls='--', lw=0.5)
plt.legend()
# plt.tight_layout() <-- REMOVE OR COMMENT THIS
# plt.show()        <-- REMOVE OR COMMENT THIS
```
